```python
import scipy.integrate
import numpy as np
from matplotlib import pyplot as plt
```

Radioactive decay

$$ \frac{dN}{dt} = -\lambda N $$

```python
def decay(t,y,lamb):
    return -lamb*y
   
# For uranium 235, half life = 703.8 million years, 5kg
lamb = np.log(2)/703.8

sol = scipy.integrate.solve_ivp(decay, [0,3000], [5], args=(lamb,),max_step=1)

# Plotting it
plt.plot(sol.t, sol.y[0])
plt.xlabel(r'time ($y\times10^6$)')
plt.ylabel('mass (kg)')
```

Predetor-prey equation (Lotka–Volterra)

$$ \frac{dx}{dt} = ax - bxy $$

$$ \frac{dy}{dt} = cxy - dy $$

```python
# defining the function
def predatorPrey(t, vec, a,b,c,d):
    x,y = vec # split the tuple y into its two values
    
    dxdt = a*x - b*x*y
    dydt = c*x*y - d*y
    
    return dxdt, dydt # return a tuple of our derivatives

# Doing the integral
sol = scipy.integrate.solve_ivp(predatorPrey, [0,50], [10,2], args=(2/3,4/3,1,1), max_step=0.1 )

# Plotting it
plt.plot(sol.t, sol.y[0], label="Prey(x)")
plt.plot(sol.t, sol.y[1], label="Predator(y)")
plt.legend()
```
