# Advanced - 1D Poisson Equation Solution

## Step 1 - Scalable poisson matrix

Let's assume we are trying to find the voltage along a 1m space with varying charge densities. 

Instead of hard coding our poisson matrix by typing out every number in the matrix, we will make one that scales to as many nodes as you want.

To do this we will use the np.diag function to make diagonal and off diagonal matricies.
[more info here](https://numpy.org/doc/stable/reference/generated/numpy.diag.html)



```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
np.diag([1])
```

```python
nodes = 100 #Number of nodes we will have in our simulation domain.
length = 1 #m, the length of our domain.
a = length/(nodes-1) # The step size 'a', in metres. 

# Have a look at the diag documentation linked above 
Poisson_Matrix = np.diag([-2]*nodes,0) + np.diag([1]*(nodes-1),1) + np.diag([1]*(nodes-1),-1)

# Print the matrix to examine it ourselves
print(Poisson_Matrix)
plt.figure()
plt.matshow(Poisson_Matrix)
```

# Step 2 - Generating our rho vector

I am going to make a $\rho$ = rho = 0.5x, where x is the distance from the centre of our system.

Try a bunch of different $\rho$ values yourself!

```python
centre = length/2
rho = np.zeros(nodes)
for i in range(nodes):
    x = i*a - centre # Find our position relative to centre
    rho[i] = 0.5*x
   
plt.figure()
plt.plot(rho)
```

```python
# This method from making the rho would have been simpler but is less generalisable to other rho functions
# centre = length/2
# np.linspace(-0.5*centre,0.5*centre,nodes)
```

## Step 3 - Adding material properties

Before we solve this, we are going to need to add in the $\frac{-a^2}{\epsilon}$ to our $\rho$.

$\epsilon$ is where we input our material properties; it's not going to change for our simulations so I will set it to be $\epsilon = 1$.


```python
eps = 1
rho_mod = rho*-1*a**2*eps
```

## Step 4 - Solve

Finally, let's generate a solution via matrix inversion. As always, this is quite a slow method (but we're lazy :) )

```python
V = np.linalg.solve(Poisson_Matrix, rho_mod)
plt.figure()
plt.plot(V)
plt.ylabel('Voltage, V')
```
