```python
#First we import a few packages we will need 
import numpy as np               # for dealing with math
import pandas as pd              # for tables
import matplotlib.pyplot as plt  # for plotting
import seaborn as sns            # for plotting
%matplotlib inline
```

```python
dataTable = pd.read_excel('CurrentVoltage.xlsx',header=2) # Setting what row to read as our heading
dataTable.columns = ["Current", "Voltage"]
dataTable
```

```python
plt.plot(dataTable.Voltage,dataTable.Current,'k.')
plt.xlabel("Volatage (V)")
plt.ylabel("Current (I)")
plt.title("Voltage vs Current")
plt.grid(True)
```

```python
x_data = dataTable.Voltage
y_data = dataTable.Current

from scipy import optimize
def test_func(x, m, b):
    return m*x + b

params, params_covariance = optimize.curve_fit(test_func, x_data, y_data, p0=[1, 0])
# give it some decent starting parameters

print("Gradient = %f, y intercept = %f" % (params[0],params[1]))

plt.figure(figsize=(15, 4)) #increase horizontal width
plt.plot(x_data, y_data,'.k' ,label='Data')

plt.plot(x_data, test_func(x_data, params[0], params[1]), "b", label='Fitted function')
plt.legend()
plt.grid()
plt.show()
```

```python
dataTable = pd.read_csv('ExpData.txt',sep='\t',header=1) # Setting what row to read as our heading
dataTable.columns = ["t", "v", "v1"]
```

```python
fixedTable = dataTable.groupby(dataTable['t'],as_index=False).aggregate({'v':'mean','v1':'mean'}) # We dont want to use t as an index
fixedTable
```

```python
meltedTable = pd.melt(fixedTable, id_vars=['t'], value_vars=['v', 'v1'])
```

```python
meltedTable
```

```python
sns.scatterplot(x="t", y="value", hue="variable", data=meltedTable)
```
